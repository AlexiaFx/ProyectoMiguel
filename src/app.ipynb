```python
import gradio as gr
from openai import OpenAI

# Conexión al servidor local de LM Studio
client = OpenAI(base_url="http://localhost:1234/v1", api_key="lm-studio")

print("✅ Conexión con LM Studio lista para modo Chat Continuo.")
```

```python
def chat_estilo_libre(mensaje_usuario, historial):
    prompt_sistema = (
        "Eres un asistente de Inteligencia Artificial avanzado, útil, detallado y amigable. "
        "Tienes la capacidad de mantener conversaciones fluidas sobre cualquier tema que el usuario proponga. "
        "Si el usuario te pregunta por materias académicas o semestres, conviértete en un asesor "
        "educativo experto y proporciónale un desglose masivo y exhaustivo de herramientas (ordenadas por categorías), "
        "materiales recomendados, libros y consejos prácticos. "
        "Sé muy abierto, detallado en tus explicaciones y mantén la continuidad del diálogo basándote en el historial."
    )
    
    mensajes = [{"role": "system", "content": prompt_sistema}]
    
    if historial:
        for interaccion in historial:
            if isinstance(interaccion, dict):
                mensajes.append({"role": "user", "content": interaccion.get("text", "")})
            elif isinstance(interaccion, (list, tuple)) and len(interaccion) >= 2:
                mensajes.append({"role": "user", "content": interaccion[0]})
                mensajes.append({"role": "assistant", "content": interaccion[1]})
        
    # Añadimos el mensaje que el usuario acaba de escribir
    mensajes.append({"role": "user", "content": mensaje_usuario})
    
    try:
        completion = client.chat.completions.create(
            model="local-model",
            messages=mensajes,
            temperature=0.7,
            max_tokens=5000 
        )
        respuesta = completion.choices[0].message.content
    except Exception as e:
        respuesta = f"❌ Error de conexión con tu IA local: {str(e)}. Verifica que LM Studio tenga el 'Local Server' encendido."
        
    return respuesta

print("✅ Cerebro conversacional de IA general configurado y optimizado.")
```
